```python
import pandas as pd
df = pd.read_csv('D_Clustered.csv')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Assuming df is your DataFrame
user_input_song = "This World"

# Retrieve audio features and cluster label of the entered song
entered_song = df[df['track_name'] == user_input_song].iloc[0]
audio_features = entered_song[['acousticness','danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness','tempo', 'valence',]]
cluster_label = entered_song['cluster_label']

# Filter DataFrame to include only songs with the same cluster label
same_cluster_songs = df[df['cluster_label'] == cluster_label]

# Calculate cosine similarity between the entered song and recommended songs
similarity_scores = cosine_similarity(audio_features.values.reshape(1, -1), same_cluster_songs[['acousticness','danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness','tempo', 'valence']])

# Add a 'similarity' column to the DataFrame
same_cluster_songs['similarity'] = similarity_scores.flatten()

# Exclude the entered song from recommendations
same_cluster_songs = same_cluster_songs[same_cluster_songs['track_name'] != user_input_song]

# Sort by similarity and get the top 5 recommendations
recommendations = same_cluster_songs.sort_values(by='similarity', ascending=False).head(10)

# Display recommendations with similarity scores
print("Top 10 recommendations: ")
print(recommendations[['track_name', 'artist_name', 'similarity']])

```
